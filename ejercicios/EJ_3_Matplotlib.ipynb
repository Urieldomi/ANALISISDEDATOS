```python
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image

imagen = Image.open('Full_Moon_Luc_Viatour.jpg')
imagen.mode
px = imagen.load()
l, a = imagen.size

#Poner los rojos en 0
for x in range(l):
    for y in range(a):
          (r,g,b) = px[x, y]
          if (r) != (0):
            px[x, y] = (0,g,b)

fig,ax = plt.subplots(figsize=(6,6))
ax.imshow(imagen)
plt.show()


#Poner los verdes en 0 
imagen2 = Image.open('Full_Moon_Luc_Viatour.jpg')
imagen2.mode
px2 =imagen2.load()

for x in range(l):
    for y in range(a):
          (r,g,b) = px2[x, y]
          if (g) != (0):
            px2[x, y] = (r,0,b)



fig,ax2 = plt.subplots(figsize=(6,6))
ax2.imshow(imagen2)
plt.show()

#Poner los azules en 0
imagen3 = Image.open('Full_Moon_Luc_Viatour.jpg')
imagen3.mode
px3 =imagen3.load()

for x in range(l):
    for y in range(a):
          (r,g,b) = px3[x, y]
          if (b) != (0):
            px3[x, y] = (r,g,0)


fig,ax3 = plt.subplots(figsize=(6,6))
ax3.imshow(imagen3)
plt.show()

```
