```python
import matplotlib.pyplot as plt
from PIL import Image
 
imagen = Image.open('Full_Moon_Luc_Viatour.jpg')
imagen.mode
px = imagen.load()
l, a = imagen.size

for x in range(l):
    for y in range(a):
          (r,g,b) = px[x, y]
          if (r, g, b) == (0,0,0):
            px[x, y] = (255, 250, 250)

fig,ax = plt.subplots(figsize=(6,6))
ax.imshow(imagen)
plt.show()
```
