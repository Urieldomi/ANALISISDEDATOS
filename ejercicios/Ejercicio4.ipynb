Leer la imagen como arreglo de numpy y transformada a escala de grises

```python
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image 

perro = Image.open('perro.jpg')
perro_a = np.array(perro)
```

```python
perrito = np.transpose(perro_a,[1,0,2])

fig,ax = plt.subplots(figsize=(6,6))

ax.imshow(perrito)

```

```python

#Poner los rojos en 0
for x in range(perrito.shape[0]):
    for y in range(perrito.shape[1]):
        g =min(perrito[x][y][0],perrito[x][y][1],perrito[x][y][2])
        perrito[x][y] =[g,g,g]
          

fig,ax = plt.subplots(figsize=(6,6))

ax.imshow(perrito)
```
